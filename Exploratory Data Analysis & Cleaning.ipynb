```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./world_population.csv')
```

```python
numeric_columns = df.select_dtypes(include=['number'])

df[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())
```

```python
df.nunique()
```

```python
df.columns
```

```python
# Univariate
numeric_columns=['Rank', '2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population', 'Area (km²)', 'Density (per km²)', 'Growth Rate',
       'World Population Percentage']
categorical_columns = ['CCA3','Country','Capital','Continent']
# histograms for numeric columns
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, start=1):
    plt.subplot(3, 5, i)
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(column)
    plt.tight_layout()

#count plots for categorical columns
plt.figure(figsize=(12, 6))
for i, column in enumerate(categorical_columns, start=1):
    plt.subplot(2, 2, i)
    sns.countplot(x=column, data=df)
    plt.title(column)
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

```python
#Bivariate
sns.pairplot(df, diag_kind='kde')

# Scatter plot
sns.scatterplot(x='Rank', y='World Population Percentage', data=df)

plt.show()
```

```python
# Box plot
sns.boxplot(x='Continent', y='Growth Rate', data=df)
```

```python
# Violin plot
sns.violinplot(x='Continent', y='2022 Population', data=df)
```

#### Correlation Coefficient Range: The values of correlation coefficients range from -1 to 1.

1: Perfect positive correlation: When one variable increases, the other variable increases proportionally.

-1: Perfect negative correlation: When one variable increases, the other variable decreases proportionally.

0: No correlation: There is no linear relationship between the two variables.

#### Interpreting the Values:

Positive Values: A positive correlation coefficient indicates a positive linear relationship between the variables. As one variable increases, the other tends to increase as well.

Negative Values: A negative correlation coefficient indicates a negative linear relationship. As one variable increases, the other tends to decrease.
